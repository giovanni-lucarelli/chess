```python
from chessrl import chess_py as cp, MCTS
from chessrl.env import LichessDefender
from chessrl.utils.plot_chess import plot_game
```

```python

g = cp.Game()
g.reset_from_fen("4k3/8/8/4K3/8/8/1R6/Q7 w - - 0 1")  # example
plot_game(g)


```

```python

# # Pure random playouts (fastest)
# mcts = MCTS(seconds=0.5, iterations=20_000, use_env_in_rollout=False)
# best = mcts.search(g)

# Or: rollouts via Env with a defender that auto-replies for Black
defender = LichessDefender()
mcts = MCTS(seconds=5.0, iterations=5_000, use_env_in_rollout=True,
            defender=defender, absorb_black_reply=True)
best = mcts.search(g)


# Or: rollouts via Env with a defender that auto-replies for Black
# defender = SyzygyDefender("/path/to/tables")
# mcts = MCTS(seconds=1.0, iterations=5_000, use_env_in_rollout=True,
#             defender=defender, absorb_black_reply=True)
# best = mcts.search(g)

if best is not None:
    print("Best move:", cp.Move.to_uci(best))
```
